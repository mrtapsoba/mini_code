## Creation de fonction de convertion de coordonnées X, Y en Longitude, latitude

```python
import utm

def convert_to_lon_lat(val_x, val_y):
    zone_number = 30
    lat, lon = utm.to_latlon(val_x, val_y, zone_number, northern=True)
    new_data = {
        'longitude': [lon],
        'latitude': [lat]
    }
    df_coorr = pd.Dataframe(new_data)
    return df_coorr
```
